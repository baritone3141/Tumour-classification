# Classification of Tumours using physical properties determined from images
These physical properties include radius, texture, perimeter, area, smoothness, compactness, etc.
Please refer to the fragment of dataset shown below for more details.

```python
import pandas as pd
import numpy as np
```

## Dataset retrieved from 
https://www.kaggle.com/datasets/yasserh/breast-cancer-dataset

```python

df = pd.read_csv('breast-cancer.csv')
df.head()
```

# Binary encoding of classes
Using 1's and 0's, it might be easier to read. Here, I used 1's to represent malignant tumours and 0's to represent benign tumours

```python
arr = []
for i in df.index:
    if df['diagnosis'][i] == "M":
        arr.append(1)
    elif df['diagnosis'][i] == "B":
        arr.append(0)
df['diagnosis_nr'] = arr
```

```python
from matplotlib import pyplot as plt
X = df.loc[:,['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean', \
              'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', \
              'fractal_dimension_mean']].to_numpy() # into numpy array
y = df.loc[:,'diagnosis_nr'].to_numpy()
```

# Split dataset

```python
from sklearn.model_selection import train_test_split

# split the data set
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.3)
```

# Models
test different models to predict whether a tumour is malignant or benign
models include logistic regression, SVM, decision tree, and random forest

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

# Logistic regression with linear features
logreg = LogisticRegression(solver = 'liblinear')
Y_train = Y_train.reshape(-1, 1).ravel()
logreg.fit(X_train, Y_train)
LRL = logreg.predict(X_test)

# Logistic regression with polynomial features
from sklearn.preprocessing import PolynomialFeatures
# degree 2
quadratic = PolynomialFeatures(degree = 2, include_bias = False)
LR = LogisticRegression(solver = 'liblinear')
X_quad = quadratic.fit_transform(X_train)
LR.fit(X_quad, Y_train)
X_test_quad = quadratic.fit_transform(X_test)
LRQ = LR.predict(X_test_quad)

#degree 3
cube = PolynomialFeatures(degree = 3, include_bias = False)
LR2 = LogisticRegression(solver = 'liblinear')
X_cube = cube.fit_transform(X_train)
LR2.fit(X_cube, Y_train)
X_test_cube = cube.fit_transform(X_test)
LRC = LR2.predict(X_test_cube)


print('Actual data:')
print(Y_test)
print()
print('Prediction using logistic regression with linear features:')
print(LRL)
print()
print('Prediction using logistic regression with polynomial (degree 2) features:')
print(LRQ)
print()
print('Prediction using logistic regression with polynomial (degree 3) features:')
print(LRC)
```

# Evaluation of some logistic regression models
Metrics used include accuracy score, F1 score, balanced accuracy score, ROC score
Also includes visualisation using a confusion matrix

```python
from sklearn.metrics import accuracy_score as acc
from sklearn.metrics import f1_score as f1
from sklearn.metrics import balanced_accuracy_score as bas
from sklearn.metrics import roc_auc_score as ras
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
print('Linear features:')
print('Accuracy score =', acc(Y_test, LRL))
print('F1 score =', f1(Y_test, LRL))
print('Balanced accuracy =', bas(Y_test, LRL))
print('ROC_AUC score =', ras(Y_test, LRL))
cmlr1 = confusion_matrix(Y_test, LRL, labels=logreg.classes_)
displr1 = ConfusionMatrixDisplay(confusion_matrix=cmlr1, display_labels=logreg.classes_)
displr1.plot()
plt.show()
print('Polynomial (degree 2) features:')
print('Accuracy score =', acc(Y_test, LRQ))
print('F1 score =', f1(Y_test, LRQ))
print('Balanced accuracy =', bas(Y_test, LRQ))
print('ROC_AUC score =', ras(Y_test, LRQ))
cmlr2 = confusion_matrix(Y_test, LRQ, labels=logreg.classes_)
displr2 = ConfusionMatrixDisplay(confusion_matrix=cmlr2, display_labels=logreg.classes_)
displr2.plot()
plt.show()
print('Polynomial (degree 3) features:')
print('Accuracy score =', acc(Y_test, LRC))
print('F1 score =', f1(Y_test, LRC))
print('Balanced accuracy =', bas(Y_test, LRC))
print('ROC_AUC score =', ras(Y_test, LRC))
cmlr3 = confusion_matrix(Y_test, LRC, labels=logreg.classes_)
displr3 = ConfusionMatrixDisplay(confusion_matrix=cmlr3, display_labels=logreg.classes_)
displr3.plot()
plt.show()
```

It appears that logistic regression with quadratic features is the better than linear and cubic

# SVM

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
# Linear SVM
clf = Pipeline((
    ("scaler", StandardScaler()),
    ("linear_svc", LinearSVC())
))
clf.fit(X_train, Y_train)
LS = clf.predict(X_test)
# Quadratic SVM
poly_clf = Pipeline((
    ("poly_features",PolynomialFeatures(degree = 2, include_bias = False)),
    ("scaler", StandardScaler()),
    ("svm_clf", LinearSVC())
))
poly_clf.fit(X_train, Y_train)
NLS = poly_clf.predict(X_test)
# Cubic SVM
poly_clf_hi = Pipeline((
    ("poly_features",PolynomialFeatures(degree = 3, include_bias = False)),
    ("scaler", StandardScaler()),
    ("svm_clf", LinearSVC())
))
poly_clf_hi.fit(X_train, Y_train)
NLS_hi = poly_clf_hi.predict(X_test)

print('Actual data:')
print(Y_test)
print()
print('Prediction using linear SVM:')
print(LS)
print()
print('Prediction using quadratic SVM:')
print(NLS)
print()
print('Prediction using cubic SVM:')
print(NLS_hi)
```

# Evaluation of several SVM models
Metrics used include accuracy score, F1 score, balanced accuracy score, ROC score
Also includes visualisation using a confusion matrix

```python
print('Linear SVM:')
print('Accuracy score =', acc(Y_test, LS))
print('F1 score =', f1(Y_test, LS))
print('Balanced accuracy =', bas(Y_test, LS))
print('ROC_AUC score =', ras(Y_test, LS))
cmsvm1 = confusion_matrix(Y_test, LS, labels=clf.classes_)
dispsvm1 = ConfusionMatrixDisplay(confusion_matrix=cmsvm1, display_labels=clf.classes_)
dispsvm1.plot()
plt.show()
print('Quadratic SVM:')
print('Accuracy score =', acc(Y_test, NLS))
print('F1 score =', f1(Y_test, NLS))
print('Balanced accuracy =', bas(Y_test, NLS))
print('ROC_AUC score =', ras(Y_test, NLS))
cmsvm2 = confusion_matrix(Y_test, NLS, labels=poly_clf.classes_)
dispsvm2 = ConfusionMatrixDisplay(confusion_matrix=cmsvm2, display_labels=poly_clf.classes_)
dispsvm2.plot()
plt.show()
print('Cubic SVM:')
print('Accuracy score =', acc(Y_test, NLS_hi))
print('F1 score =', f1(Y_test, NLS_hi))
print('Balanced accuracy =', bas(Y_test, NLS_hi))
print('ROC_AUC score =', ras(Y_test, NLS_hi))
cmsvm3 = confusion_matrix(Y_test, NLS_hi, labels=poly_clf_hi.classes_)
dispsvm3 = ConfusionMatrixDisplay(confusion_matrix=cmsvm3, display_labels=poly_clf_hi.classes_)
dispsvm3.plot()
plt.show()
```

For SVM, it appears that non-linear is better than linear
But degree 2 is slightly better than degree 3

# Decision Tree

```python
# Decision tree
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier(max_depth=2)
dt_clf.fit(X_train, Y_train)
DT = dt_clf.predict(X_test)

print('Actual data:')
print(Y_test)
print()
print('Prediction using Decision tree:')
print(DT)
print()
```

# Evaluation of a decision tree model
Metrics used include accuracy score, F1 score, balanced accuracy score, ROC score
Also includes visualisation using a confusion matrix

```python
# from decision tree,
print('Accuracy score =', acc(Y_test, DT))
print('F1 score =', f1(Y_test, DT))
print('Balanced accuracy =', bas(Y_test, DT))
print('ROC_AUC score =', ras(Y_test, DT))

cmdt = confusion_matrix(Y_test, DT, labels=dt_clf.classes_)
dispdt = ConfusionMatrixDisplay(confusion_matrix=cmdt, display_labels=dt_clf.classes_)
dispdt.plot()
plt.show()
```

# Random Forest

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier()
rf_clf.fit(X_train, Y_train)
RF = rf_clf.predict(X_test)

print('Actual data:')
print(Y_test)
print()
print('Prediction using Decision tree:')
print(RF)
print()
```
